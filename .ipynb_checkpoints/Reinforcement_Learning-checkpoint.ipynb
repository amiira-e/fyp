## Outliers

```python
from scipy.stats import trim_mean

# Define the trimming fraction
trim_fraction = 0.1

# Calculate the trimmed mean for each column
trimmed_means = df_sample.apply(lambda x: trim_mean(x, trim_fraction))

# Replace the outliers with the trimmed mean
df_sample = df_sample.apply(lambda x: np.where(np.abs(x - np.mean(x)) > 2*np.std(x), trimmed_means[x.name], x))
```

## Class imbalance

In the context of network analysis, one way to handle class imbalance is to adjust the weights of the edges in the network. For example, you can assign higher weights to edges that correspond to fraudulent activities to increase their impact on the overall network structure

## Network building

```python
# preprocess the data as needed
G = nx.from_pandas_edgelist(df, source='nameOrig', target='nameDest', edge_attr='amount')
```

```python
# define nodes and edges
G.add_nodes_from(df['nameOrig'].unique(), bipartite=0)
G.add_nodes_from(df['nameDest'].unique(), bipartite=1)
G.add_weighted_edges_from(df[['nameOrig', 'nameDest', 'amount']].values)
```

Analyze the network: Use network analysis algorithms to identify patterns and anomalies in the network that may be indicative of fraud. For example, you could use centrality measures to identify nodes that are more important or influential in the network, or use community detection algorithms to identify clusters of nodes that are more tightly connected than expected.

```python
# compute centrality measures
degree_centrality = nx.degree_centrality(G)
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)

# compute community detection
communities = nx.algorithms.community.greedy_modularity_communities(G)
```

Integrate with other fraud detection techniques: Use the results of your network analysis to inform other fraud detection techniques, such as rule-based systems or machine learning models. For example, you could use the insights from network analysis to create rules for flagging suspicious transactions or to train a machine learning model to detect fraud.

```python
# use network analysis to flag suspicious transactions
fraud_threshold = 0.8
suspicious_transactions = []
for u, v, d in G.edges(data=True):
    if degree_centrality[u] > fraud_threshold or degree_centrality[v] > fraud_threshold:
        suspicious_transactions.append((u, v, d['amount']))
```

```python
import networkx as nx

G = nx.DiGraph()
```

```python
G.add_nodes_from(df_sample['nameOrig'].unique())
G.add_nodes_from(df_sample['nameDest'].unique())
```

```python
for index, row in df_sample.iterrows():
    G.add_edge(row['nameOrig'], row['nameDest'], weight=row['amount'])
```

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'amount')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```
